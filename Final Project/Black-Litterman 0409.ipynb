```python
etf=etf.dropna()
etf_price=etf.iloc[np.where(etf['DATES']>'2013-12-30')[0],:].reset_index(drop=True).set_index('DATES')
etf_RET=etf_price.pct_change().dropna()
etf_C_RET=(etf_RET+1).cumprod()-1
# Assume Rf=2% annualized
rf=0.02
rf_day=(1+rf)**(1/252)-1
etf_excess_RET=etf_RET-rf_day
etf_C_excessRET=(etf_excess_RET+1).cumprod()-1
etf_C_excessRET
```

```python
# Calculate for weights
etf_mktcap=mktcap.iloc[np.where(mktcap['DATES']>'2013-12-31')[0],:].reset_index(drop=True).set_index('DATES')
etf_mktcap=etf_mktcap.dropna()
etf_weights=etf_mktcap.copy()
for i in range(etf_mktcap.shape[0]):
    etf_weights.iloc[i,:]=etf_weights.iloc[i,:]/np.sum(etf_mktcap,axis=1)[i]
```

## Column 3&4 -

```python
'''
Implied Excess Equilibrium Return = lambda*sigma*weights
lambda: risk aversion coefficient. We can use 'Sharpe ratio' to represent it. (Scalar)
sigma: covariance matrix (NxN)
weights: market capitalization weight (Nx1)
'''
```

```python
temp=pd.Series(np.sum(etf_RET.iloc[:]*etf_weights.iloc[:],axis=1),name='value weighted portfolio')
VWP=pd.DataFrame(data=temp)
rf=0.02
rf_day=(1+rf)**(1/252)-1
Lambda=(np.mean(VWP.iloc[:,0])-rf_day)/np.var(VWP.iloc[:,0])
```

```python
Lambda
```

```python
# Based on investopedia 
cov_matrix=(etf_RET*100).cov()
cov_matrix
```

```python
weights=etf_weights.iloc[-1:,:]
weights
```

```python
Col1_4=Lambda*np.dot(cov_matrix,weights.T)
pd.DataFrame(data=Col1_4.T,columns=b)
# 這邊單位是(%)
```

```python
'''
Inverse method:
weights=(lambda*sigma)^-1*Implied Excess Equilibrium Return
lambda: risk aversion coefficient. We can use 'Sharpe ratio' to represent it. (Scalar)
sigma: covariance matrix (NxN)
weights: market capitalization weight (Nx1)
'''
```
